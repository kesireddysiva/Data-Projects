## Univariate Analysis

1.   Distribution of Athletes by Gender(sex).
2.   Display the top 10 most popular sports based on athletes participation.
3.   Find the most popular sport based on the participation.
4.   check the age distribution.







```python
# Distribution of Athletes by Gender(sex).
gen_dist=df.groupby('Sex',as_index=False)['ID'].count().rename(columns={'ID':'Count'})
plt.figure(figsize=(8,6))
sns.barplot(x='Sex',y='Count',data=gen_dist)
plt.title('Distribution of Athletes by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
# Display the top 10 most popular sports based on athletes participation

top_10=df['Sport'].value_counts().head(10)
plt.figure(figsize=(8,6))
sns.barplot(y=top_10.index,x=top_10.values)
plt.title('Top 10 Most Popular Sports')
plt.xlabel('Sport')
plt.ylabel('Count')
plt.show()
```

```python
df['Sport'].value_counts().head(10).plot(kind='pie',autopct='%1.2f%%',figsize=(8,6),startangle=90)
```

```python
# Find the most popular sport based on the participation.
most_popular=df['Sport'].value_counts().idxmax()
print('Most Popular Sport:',most_popular)
```

```python
#check the age ditribution.
sns.histplot(df['Age'],bins=20,kde=True)
plt.title("Age Distribution")
plt.show()
```

```python
sns.boxplot(x=df['Age'])
```

### Bivariate and Multivariate Analysis

```python
# Numeric-Numeric relationship
sns.scatterplot(x='Age',y='Height',data=df)
plt.title("Age Vs Height")
plt.show()
```

```python
sns.scatterplot(x='Age',y='Height',hue='Sex',data=df)
```

```python
# Numeric-Categorical Relationship(Age distribution by medal)
sns.barplot(x='Medal',y='Age',data=df)
```

```python
# How many medals Gold, Sliver, Bronze medals have been awarded to female and male athlete?
dff=df.groupby(['Medal','Sex'],as_index=False).size()
pivot_df = dff.pivot(index="Sex", columns="Medal", values="size")
pivot_df
```

```python


# Plot a grouped bar chart
pivot_df.plot(kind="bar", figsize=(8, 6))

# Customize the plot
plt.title("Medal Count by Gender", fontsize=16)
plt.xlabel("Gender", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.xticks(rotation=0)
plt.legend(title="Medal Type", fontsize=10)
plt.grid(axis="y", linestyle="--", alpha=0.7)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
sns.heatmap(pivot_df)
```

```python
#display no.of.medal year wise for india.
# Display the number of medals won by India year-wise
df_india = df[df['Team'].str.lower() == "india"]
df_india = df_india.groupby('Year', as_index=False)['Medal'].count().rename(columns={"Medal": "Count"})
df_india

# Plot the year-wise medal count
sns.lineplot(x='Year', y='Count', data=df_india)
plt.title("Year-wise Medal Count for India")
plt.xlabel("Year")
plt.ylabel("Medal Count")
plt.show()
```

```python
# Display the top 10 athletes by medal count
top_10 = df['Name'].value_counts().head(10)

# Plot the top 10 athletes
sns.barplot(x=top_10.values, y=top_10.index)
plt.title('Top 10 Athletes by Medal Count')
plt.xlabel('Medal Count')
plt.ylabel('Athlete')
plt.show()
```
