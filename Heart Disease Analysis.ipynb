```python
# 12. Find Gender Distribution According to The Target Variable

df_filtered = df.groupby(['sex', 'target'], as_index=False).size()

# Visualize the gender distribution according to the target variable
sns.barplot(data=df_filtered, x='sex', y='size', hue='target')
plt.title("Gender Distribution According to the Target Variable")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```
